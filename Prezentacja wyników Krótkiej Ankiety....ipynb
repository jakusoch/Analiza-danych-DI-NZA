```python
df['Czy systemowo informujecie każdą osobę na waszej OU o Deklaracji Ideowej?'].value_counts()
```

```python
data=df['Czy systemowo informujecie każdą osobę na waszej OU o Deklaracji Ideowej?'].value_counts()
labels = data.index.tolist()

colors = sns.color_palette('colorblind')[0:5]

#create pie chart
plt.figure(figsize=(8,6))
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')

plt.title('Wstępna ankieta do Przewodniczących: \nPytanie 4. Czy systemowo informujecie każdą osobę na waszej OU o Deklaracji Ideowej?',fontsize=16,x=-0.3,ha="left")
plt.show()
```

```python
data=df['Jak dużo osób w twojej OU wie o istnieniu Deklaracji Ideowej?'].value_counts()
data = pd.DataFrame(data.items(), columns=['odp', 'Count'])
nowy_wiersz = {'odp': 'Wszyscy lub niemal wszyscy', 'Count': 0}
data = data.append(nowy_wiersz, ignore_index=True)
print(data)
data['odp'][0]='Jak wyżej + pojedyncze \nosoby zainteresowane tematem'

labels = data['odp']


colors = sns.color_palette('colorblind')[0:5]

#create pie chart
plt.figure(figsize=(8,6))
plt.pie(data['Count'], labels = labels, colors = colors, autopct='%.0f%%')
#plt.legend(loc='upper right')
plt.title('Wstępna ankieta do Przewodniczących: \nPytanie 5. Jak dużo osób w twojej OU wie o istnieniu Deklaracji Ideowej?\n(pytanie zadane tylko przy odpowiedzi "Nie" lub "Inne" w poprzednim pytaniu)',fontsize=16,x=-0.25,ha="left")
plt.show()
```

```python
data=df['Jak najczęściej jest przedstawiana Deklaracja Ideowa na twojej OU?'].value_counts()

data = pd.DataFrame(data.items(), columns=['odp', 'Count'])

nowy_wiersz = {'odp': 'Jest to obowiązujący dokument, \nktórym musimy się kierować', 'Count': 0}
data = data.append(nowy_wiersz, ignore_index=True)
data['odp'][0]='Jest to obowiązujący dokument, ale się \nnim nie kierujemy (nie musimy kierować)'
data['odp'][1]='Może i teoretycznie jest w statucie, \nale my uznajemy, że nas nie obowiązuje'
data['odp'][2]='Jest to dokument, który kiedyś \nobowiązywał ale już nie obowiązuje'
labels = data['odp']


colors = sns.color_palette('colorblind')[0:5]

#create pie chart
plt.figure(figsize=(8,6))
plt.pie(data['Count'],labels=labels, colors = colors, autopct='%.0f%%')
legend_labels = [f"{label}: {count}" for label, count in zip(labels, data)]
#plt.legend(legend_labels, loc="best")
plt.title('Wstępna ankieta do Przewodniczących: \nPytanie 6. Jak najczęściej jest przedstawiana Deklaracja Ideowa na twojej OU?',fontsize=16,x=-0.44,ha="left")
plt.show()
```
